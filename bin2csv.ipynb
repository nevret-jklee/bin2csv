## Bin file to CSV file

```python
import numpy as np
import pandas as pd

magic, seq = [], []
x_acc, y_acc, z_acc = [], [], []
roll, patch, yaw = [], [], []
sound, loadcell, pressure, pos_cmd, pos_fdb = [], [], [], [], []
limit_sw, temp, humid, dev_err, rms = [], [], [], [], []

# columns = [magic, seq, x_acc, y_acc, z_acc, roll, patch, yaw, sound, loadcell, pressure, pos_cmd, pos_fdb, limit_sw, temp, humid, dev_err, rms]

with open('20200825-1-Header.bin','r', errors='ignore') as f:
    '''
    ndarray 데이터 형태 지정: https://rfriend.tistory.com/tag/dtype
    
    '''
    for _ in range(11906): # len(bin file)
        magic.append(np.fromfile(f, '<u2', 2))        # uint32, little endian
        seq.append(np.fromfile(f, '<u2', 1))          # uint16, little endian
        x_acc.append(np.fromfile(f, '<i2', 1))        # int16, little endian
        y_acc.append(np.fromfile(f, '<i2', 1))        # int16, little endian
        z_acc.append(np.fromfile(f, '<i2', 1))        # int16, little endian
        roll.append(np.fromfile(f, '<i2', 1))         # int16, little endian
        patch.append(np.fromfile(f, '<i2', 1))        # int16, little endian
        yaw.append(np.fromfile(f, '<i2', 1))          # int16, little endian
        sound.append(np.fromfile(f, '<u2', 1))        # uint16, little endian
        loadcell.append(np.fromfile(f, '<i2', 1))     # int16, little endian
        pressure.append(np.fromfile(f, '<u1', 1))     # uint8, little endian
        pos_cmd.append(np.fromfile(f, '<u1', 1))      # uint8, little endian
        pos_fdb.append(np.fromfile(f, '<u1', 1))      # uint8, little endian
        limit_sw.append(np.fromfile(f, '<u1', 1))     # uint8, little endian
        temp.append(np.fromfile(f, '<i1', 1))         # int8, little endian
        humid.append(np.fromfile(f, '<u1', 1))        # uint8, little endian
        dev_err.append(np.fromfile(f, '<u2', 1))      # uint16, little endian
        rms.append(np.fromfile(f, '<u2', 1))          # uint16, little endian
```

### list to numpy array & reshape array

```python
seq = np.array(seq).reshape(-1, )
x_acc = np.array(x_acc).reshape(-1, )
y_acc = np.array(y_acc).reshape(-1, )
z_acc = np.array(z_acc).reshape(-1, )
roll = np.array(roll).reshape(-1, )
patch = np.array(patch).reshape(-1, )
yaw = np.array(yaw).reshape(-1, )
sound = np.array(sound).reshape(-1, )
loadcell = np.array(loadcell).reshape(-1, )
pressure = np.array(pressure).reshape(-1, )
pos_cmd = np.array(pos_cmd).reshape(-1, )
pos_fdb = np.array(pos_fdb).reshape(-1, )
limit_sw = np.array(limit_sw).reshape(-1, )
temp = np.array(temp).reshape(-1, )
humid = np.array(humid).reshape(-1, )
dev_err = np.array(dev_err).reshape(-1, )
rms = np.array(rms).reshape(-1, )
```

### check array

```python
print(seq)
print(x_acc)
print(y_acc)
print(z_acc)
print(roll)
print(patch)
print(yaw)
print(sound)
print(loadcell)
print(pressure)
print(pos_cmd)
print(pos_fdb)
print(limit_sw)
print(temp)
print(humid)
print(dev_err)
print(rms)
```

### concat array and transpose

```python
data = np.array([seq, x_acc, y_acc, z_acc,
                 roll, patch, yaw,
                 sound, loadcell, pressure, pos_cmd, pos_fdb,
                 limit_sw, temp, humid, dev_err, rms])
data = data.T
data
```

### make array to dataframe

```python
columns = ['seq', 'x_acc', 'y_acc', 'z_acc', 
           'roll', 'patch', 'yaw', 
           'sound', 'loadcell', 'pressure', 'pos_cmd', 'pos_fdb', 
           'limit_sw', 'temp', 'humid', 'dev_err', 'rms']

df = pd.DataFrame(data = data,
                  columns = columns)
df
```

```python
df.to_csv('bin2csvfile.csv',
          index=False)
```
